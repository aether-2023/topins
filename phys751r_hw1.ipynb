```python
import numpy as np
import matplotlib.pyplot as plt

k = np.linspace(-np.pi, np.pi, 1000)

# Insulating Phase 1:   0
t_ss1, t_pp1, t_sp1 = 1, 0.5, 0.3
epsilon1 = t_ss1 - t_pp1
E_plus1 = -epsilon1 * np.cos(k) + np.sqrt((t_ss1 + t_pp1)**2 * np.cos(k)**2 + 4 * t_sp1**2 * np.sin(k)**2)
E_minus1 = -epsilon1 * np.cos(k) - np.sqrt((t_ss1 + t_pp1)**2 * np.cos(k)**2 + 4 * t_sp1**2 * np.sin(k)**2)

# Insulating Phase 2: t_sp = 0
t_ss2, t_pp2, t_sp2 = 1, 1, 0
epsilon2 = t_ss2 - t_pp2
E_plus2 = -epsilon2 * np.cos(k) + np.sqrt((t_ss2 + t_pp2)**2 * np.cos(k)**2 + 4 * t_sp2**2 * np.sin(k)**2)
E_minus2 = -epsilon2 * np.cos(k) - np.sqrt((t_ss2 + t_pp2)**2 * np.cos(k)**2 + 4 * t_sp2**2 * np.sin(k)**2)

# Transition Point:  = 0, t_sp  0
t_ss3, t_pp3, t_sp3 = 0.7, 0.3, 0.2
epsilon3 = t_ss3 - t_pp3
E_plus3 = -epsilon3 * np.cos(k) + np.sqrt((t_ss3 + t_pp3)**2 * np.cos(k)**2 + 4 * t_sp3**2 * np.sin(k)**2)
E_minus3 = -epsilon3 * np.cos(k) - np.sqrt((t_ss3 + t_pp3)**2 * np.cos(k)**2 + 4 * t_sp3**2 * np.sin(k)**2)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

axs[0].plot(k, E_plus1, 'b', label='$E_+$')
axs[0].plot(k, E_minus1, 'r', label='$E_-$')
axs[0].set_title('Insulating Phase 1 ($\\epsilon \\neq 0$)')
axs[0].set_xlabel('$k$')
axs[0].set_ylabel('$E(k)$')
axs[0].legend()

axs[1].plot(k, E_plus2, 'b', label='$E_+$')
axs[1].plot(k, E_minus2, 'r', label='$E_-$')
axs[1].set_title('Insulating Phase 2 ($t_{sp} = 0$)')
axs[1].set_xlabel('$k$')

# axs[2].plot(k, E_plus3, 'b', label='$E_+$')
# axs[2].plot(k, E_minus3, 'r', label='$E_-$')
# axs[2].set_title('Transition Point ($\\epsilon = 0$)')
# axs[2].set_xlabel('$k$')

plt.tight_layout()
plt.show()
```
